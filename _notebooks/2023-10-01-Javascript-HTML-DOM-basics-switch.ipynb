```python
%%html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Grocery List</title>
</head>
<body>
    <!-- Grocery List Title -->
    <h1 id="storeTitle">Grocery List</h1>

    <!-- Grocery List -->
    <ul id="groceryList">
        <li>Apples</li>
        <li>Bananas</li>
        <li>Carrots</li>
        <li>Oranges</li>
        <li>Milk</li>
    </ul>

    <!-- Swap Items Form -->
    <div>
        <input type="text" id="item1Input" placeholder="Item 1">
        <input type="text" id="item2Input" placeholder="Item 2">
        <button id="swapItemsButton">Switch Items</button>
    </div>

    <!-- JavaScript Code -->
    <script>
        // Function to swap two list items
        function swapItems() {
            var item1Input = document.getElementById("item1Input").value;
            var item2Input = document.getElementById("item2Input").value;
            var groceryList = document.getElementById("groceryList");

            var items = groceryList.getElementsByTagName("li");

            var item1, item2;

            for (var i = 0; i < items.length; i++) {
                if (items[i].innerText === item1Input) {
                    item1 = items[i];
                }
                if (items[i].innerText === item2Input) {
                    item2 = items[i];
                }
            }

            if (item1 && item2) {
                var item1Index = Array.from(items).indexOf(item1);
                var item2Index = Array.from(items).indexOf(item2);

                // Swap the items
                groceryList.insertBefore(item1, items[item2Index]);
                groceryList.insertBefore(item2, items[item1Index]);

                // Update the title
                var storeTitle = document.getElementById("storeTitle");
                storeTitle.innerText = "Grocery List (Switched)";
            }
        }

        // Attach the swapItems function to the button click event
        var swapItemsButton = document.getElementById("swapItemsButton");
        swapItemsButton.onclick = swapItems;
    </script>
</body>
</html>
```

### Utterance

*When the "Swap Items" button is clicked, it swaps the positions of two items in the grocery list that have been entered and updates the title to "Grocery List (Switched)"*
